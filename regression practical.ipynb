## Independent and dependent feature 

```python
dataset.head()
```

```python
X=dataset.iloc[:,:-1]
y = dataset.iloc[:,-1]
```

```python
X.head()
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
#xtest and ytest is used when model is build
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
y_test
```

##  Standardize or feature scaling the datasets

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
#it is set to transform to avoid data leakage
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

## Model training

# linear regression

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
```

```python
regression
```

```python
# xtrain is an independent feature and y train is an dependent feature
regression.fit(X_train,y_train)
```

```python
##PRINT THE COEFFICIENT AND THE INTERCEPT
print(regression.coef_)
```

```python
print(regression.intercept_)
```

## Prediction for test data

```python
reg_pred = regression.predict(X_test)
```

```python
reg_pred
```

## Assumption of linear regression

```python
plt.scatter(y_test,reg_pred)
plt.xlabel("test truth data")
plt.ylabel("test predicted data")
```

## residuals

```python
residuals = y_test-reg_pred
```

```python
residuals
```

```python
sns.displot(residuals,kind='kde')
```

## scatter plot with predictions and residuals
### uniform distribution

```python
plt.scatter(reg_pred,residuals)
```

## performance metrics

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
print(mean_squared_error(y_test,reg_pred))
print(mean_absolute_error(y_test,reg_pred))
print(np.sqrt(mean_squared_error(y_test,reg_pred)))
```

```python
## r Square and adj r square
from sklearn.metrics import r2_score
score = r2_score(y_test,reg_pred)
print(score)
```

```python
## adjusted r squared 
### display asjusted r squared 
1-(1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```
